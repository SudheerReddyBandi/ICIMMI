```python
corrmat = df1.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(28,28))
#Plotting heat map
g=sns.heatmap(df1[top_corr_features].corr(),annot=True,linewidths=.5)
b, t = plt.ylim()
b += 0.5
t -= 0.5
plt.ylim(b, t)
plt.title('Correlation Matrix',fontdict = {'fontsize' : 20})
plt.show()
```

```python
# Splitting the dataset into features and outcome variable(Overall)
y = df1.iloc[:,33].values
# Picking up the top ten most useful features based on the above heatmap
x = df1[['centroid_x','centroid_y','c_percent WFH','c_percent 2. Professional occupations','c_percent 5. Skilled trades occupations','c_percent 6. Caring leisure and other service occupations','c_percent very good health','c_percent good health','c_percent fair health','c_percent bad health','e_NO2','e_ozone','o_opioids_quantity_per_capita','o_hypertension_quantity_per_capita','o_depression_quantity_per_capita'
            ]].values
print(x)
```

```python
print(y)
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
## Training the multiple linear regression on the training set
from sklearn.linear_model import LinearRegression
regressor_MultiLinear = LinearRegression()
MLR=regressor_MultiLinear.fit(x_train,y_train)
```

```python
## Predicting test results
y_pred = regressor_MultiLinear.predict(x_test)
```

```python
print(y_pred)
```

```python
# Calculating r2 score
from sklearn.metrics import r2_score
r2_MultiLinear = r2_score(y_test,y_pred)
print(r2_MultiLinear)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
error = (y_test - y_pred)**2
# calculate errors
Mean_Squared_Error = mean_squared_error(y_test, y_pred)
# calculate error
Absolute_error = abs((y_test - y_pred))
```

```python
from statistics import mean
print(mean(error))
print(Mean_Squared_Error)
print(mean(Absolute_error))
```

```python
## Finding out the optimal degree of polynomial regression
from sklearn.preprocessing import PolynomialFeatures
sns.set_style('darkgrid')
scores_list = []
pRange = range(2,6)
for i in pRange :
    poly_reg = PolynomialFeatures(degree=i)
    x_poly = poly_reg.fit_transform(x_train)
    poly_regressor = LinearRegression()
    poly_regressor.fit(x_poly,y_train)
    y_pred = poly_regressor.predict(poly_reg.fit_transform(x_test))
    scores_list.append(r2_score(y_test,y_pred))
plt.plot(pRange,scores_list,linewidth=2)
plt.xlabel('Degree of polynomial')
plt.ylabel('r2 score with varying degrees')
plt.show()
```
